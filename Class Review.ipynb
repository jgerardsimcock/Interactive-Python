```python
##Euclids Algorithm for finding the GCD

def gcd(m,n):
    
    while m%n != 0:
        #this bit of logic where we create and assign...
        oldm = m
        oldn = n
        ##is different from this where we reassign
        m = oldn
        n = oldm%oldn
        
    return n
    
class Fraction:
    
    ##Python Classes always begin with a constructor function that initializes themselves
    ##We need to represent and model the data somehow
    ##In this case we need a top and bottom to represent the numerator and denomitor
    
    def __init__(self, top, bottom):
        
        self.num = top #This creates the num data object and assigns its value to the parameter top
        self.den = bottom #this creates the den data object and assigns its value to the parameter bottom
        
        ##myFraction = Fraction(3,5) will create a fraction object with a state value of 3/5 
        
    ##If we want to give more methods we need to describe how the methods will work
    
    def show(self):
        #to print we need to tell the object how to represent itself as a string
        
        print(self.num,'/', self.den)
        
    def __str__(self):
        ##There is a built-in to string method. We will override it to serve our specific needs
        return str(self.num) + "/" + str(self.den)
    
    def __add__(self, otherFraction):
        #We need to override the addition method when using the Fraction class
        ##To add fractions we need the same denominator
        
        newnum = self.num*otherFraction.den + self.den*otherFraction.num
        newdenom = self.den*otherFraction.den
        
        ##We'll bring in the gcd function to reduce the fraction to its lowest common denominator
        
        common = gcd(newnum, newdenom)
        
        return Fraction(newnum//common, newdenom//common)
    
    def __sub__(self, otherFraction):
        
        newnum = self.num*otherFraction*den - otherFraction.num*self.den
        newdenom = self.den*otherFraction.den
        
        common = gcd(newnum, newdenom)
        
        return Fraction(newnum//common,newdenom//common)
    
    def __mult__(self,otherFraction):
        
        newnum = self.num*otherFraction.num
        newdenom = self.den*otherFraction.den
        
        common = gcd(newnum, newdenom)
        
        return Fraction(newnum//common, newdenom//common)
    
    def __truediv__(self, otherFraction):
        
        newnum = self.num*otherFraction.den
        newdenom = self.den*otherFraction.num
        
        common = gcd(newnum, newdenom)
        
        return Fraction(newnum//common, newdenom//common)
    

    
    def __eq__(self, other):
        
        ##We need to override the equality method to implement deep equality
        ##__eq__ can only compare objects that point to the same reference object
        ##We want to be able to compare the values
        
        firstnumerator = self.num*other.den
        secondnumerator = other.num*self.den
        
        return firstnumerator == secondnumerator 
    
    def getNum(self):
        return self.num
    
    def getDen(self):
        return self.den
    
```

```python
class LogicGate:
    
    def __init__(self,n):
        
        ##Every gate has a label and an output this is the state of the data
        self.label = n
        self.output = None
        
        ##We need to be able to run a method to get the label 
    def getLabel(self):

        return self.label

    ##We need to be able to run a method to know our output
    def getOutput(self):
        ##Since output is based on a logic that we have defined through out binary/unary gate logics 
        ##We can run a logic method to get our output.
        ##We need to define this logic later
        self.output = self.performGateLogic()
        return self.output
        
class BinaryGate(LogicGate):
    ##BinaryGate is a sublcass of LogicGate. It inherits the methods of that class
    
    def __init__(self, n):
        LogicGate.__init__(self,n)
        
        self.PinA = None
        self.PinB = None
        
        
    def getPinA(self):
        if self.PinA == None:
            return int(input("Enter Pin A input for gate "+ self.getLabel()+"-->"))
        else: 
            return self.pinA.getFrom().getOutput()

    def getPinB(self):
        if self.PinB == None:
            return int(input("Enter Pin B input for gate "+ self.getLabel()+"-->"))
        else: 
            return self.pinB.getFrom().getOutput()
    
    def setNextPin(self, source):
        
        if self.PinA == None:
            self.PinA == source
        else: 
            if self.PinB == None:
                self.PinB == source
            else: 
                raise RuntimeError("No Open Pins")

            
class AndGate(BinaryGate):
    
    def __init__(self,n):
        
        BinaryGate.__init__(self,n)
        
    def performGateLogic(self):

        a = self.getPinA()
        b = self.getPinB()

        if a == 1 and b == 1:
            return 1
        else:
            return 0
        
class NandGate(BinaryGate):
    
    def __init__(self,n):
        BinaryGate.__init__(self,n)
 
    
    def performGateLogic(self):
        
        a = self.getPinA()
        b = self.getPinB()
        return NotGate(AndGate(self))
        
class NorGate(BinaryGate):
    
    def __init__(self,n):
        
        BinaryGate.__init__(self,n)
        
    def performGateLogic(self):
        
        a = self.getPinA()
        b = self.getPinB()
        return NotGate(OrGate(self))

class XorGate(BinaryGate):
    
    def __init__(self,n):
        
        BinaryGate.__init(self,n)
        
    def performGateLogic(self):
        
        if a == 1 and b == 1:
            return 0
        elif (a ==1 and b == 0) or (a== 0 or b == 1):
            return 1
        else: 
            return 0
            
class OrGate(BinaryGate):
    
    def __init__(self,n):
        
        BinaryGate.__init__(self,n)
        
    def performGateLogic(self):
        
        a = self.getPinA()
        b = self.getPinB()
        
        if a == 1 or b == 1:
            return 1
        else: 
            return 0
        

class UnaryGate(LogicGate):
    
    def __init__(self,n):
        LogicGate.__init__(self,n)
        
        self.pin = None
        
    def getPin(self):
        if self.pin == None:
            return int(input("Enter Pin input for gate "+ self.getLabel()+"-->"))
        else: 
            return self.pin.getFrom().getOutput()

    
    def setNextPin(self,source):
        
        if self.pin == None:
            self.pin == source
        else: 
            print "No empty Pins"

class NotGate(UnaryGate):
    
    def __init__(self,n):

        UnaryGate.__init__(self,n)

    def performGateLogic(self):

        

        if self.getPin():
            return 0
        else: 
            return 1
        
class Connector:

    def __init__(self, fgate, tgate):
        self.fromgate = fgate
        self.togate = tgate

        tgate.setNextPin(self)

    def getFrom(self):
        return self.fromgate

    def getTo(self):
        return self.togate
        
        
```
